```python
# Importing Keras library from TensorFlow
from tensorflow import keras

# Import train_test_split
from sklearn.model_selection import train_test_split
```

```python
#load the data in from CIFAR10
(train_images, train_labels), (test_images, test_labels) = keras.datasets.cifar10.load_data()
```

```python
print("Shape of train_images:", train_images.shape)
print("Shape of train_labels:", train_labels.shape)
print("Shape of test_images:", test_images.shape)
print("Shape of test_labels:", test_labels.shape)
```

SPLITTING THE TRAINING DATA.

```python
#structure from previous labs
train_images, validation_images, train_labels, validation_labels = train_test_split(
    train_images, train_labels, test_size=0.2, random_state=42)
```

```python
validation_images.shape
```

```python
train_images.shape
```

```python
train_images, test_images = train_images / 255., test_images / 255.
```

```python
#Normalizing 
validation_images = validation_images / 255.
```

```python
model = keras.Sequential([
    #The first layer is 32 filter by 3x3. Input is 32x32 with 3 colors.
    keras.layers.Conv2D(32, kernel_size=(3,3) ,  activation="relu", input_shape=(32, 32, 3)),

    keras.layers.Conv2D(64, kernel_size=(3,3), activation="relu"),
    
    # Using MaxPooling2D I did a 2x2 pool size
    keras.layers.MaxPooling2D((2, 2)),
    
   
    keras.layers.MaxPooling2D((2, 2)),
    
    keras.layers.Flatten(),
    
    #dense layer with 64 units and a relu activation
    keras.layers.Dense(64, activation="relu"),
    
    
    keras.layers.Dense(10)
])
```

COMPILING THE MODEL

```python
#https://github.com/christianversloot/machine-learning-articles/blob/main/how-to-use-sparse-categorical-crossentropy-in-keras.md
#using this github page I was able to find a better compiler
model.compile(optimizer="adam",
              loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
hist = model.fit(train_images, train_labels, validation_data=(validation_images, validation_labels), epochs=20, batch_size=300)
```

PRINTING THE RESULTS

```python
acc_score = model.evaluate(test_images, test_labels)
print("Accuracy: ", acc_score[1])
```

By using the model.compiler christianversloot recommended : https://github.com/christianversloot/machine-learning-articles/blob/main/how-to-use-sparse-categorical-crossentropy-in-keras.md I was able to achieve a decent accuracy. The accuracy I got was 66.76%
